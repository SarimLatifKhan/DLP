```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim.lr_scheduler import ReduceLROnPlateau
import torchvision.models as models
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
import torchvision.datasets as datasets

# Assuming you have the dataset prepared in 'trainmerged' and 'testmerged' folders

# Define transformations for the training and validation sets
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Resize images to 224x224
    transforms.ToTensor(),           # Convert images to tensors
    transforms.Normalize(mean=[0.485], std=[0.229])  # Normalize
])

# Load the dataset
train_dataset = datasets.ImageFolder(root='/kaggle/working/trainmerged', transform=transform)
test_dataset = datasets.ImageFolder(root='/kaggle/working/testmerged', transform=transform)

# Define data loaders
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# Define the ResNet model with regularization
resnet = models.resnet18(pretrained=True)
for param in resnet.parameters():
    param.requires_grad = False

# Replace the final fully connected layer for 7 classes
num_ftrs = resnet.fc.in_features
resnet.fc = nn.Sequential(
    nn.Linear(num_ftrs, 256),
    nn.ReLU(),
    nn.Dropout(0.5),  # Dropout layer for regularization
    nn.Linear(256, 7)
)

# Move model to GPU if available
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
resnet.to(device)

# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(resnet.parameters(), lr=0.0001)

# Learning rate scheduling
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=5, verbose=True)

# Training loop with early stopping
num_epochs = 32
batch_size = 128
best_val_loss = float('inf')
model.summary()
```

```python
import time

avg_epoch_time = 0  # Initialize average epoch time
total_steps = len(train_loader)  # Total number of steps per epoch
for epoch in range(num_epochs):
    start_time = time.time()  # Start timing for the epoch
    resnet.train()
    running_loss = 0.0
    correct = 0
    total = 0
    for i, data in enumerate(train_loader, 0):
        inputs, labels = data
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()

        outputs = resnet(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

        # Calculate progress
        progress = (i + 1) / total_steps
        # Calculate estimated time remaining for epoch
        time_elapsed = time.time() - start_time
        time_per_step = time_elapsed / (i + 1)
        steps_remaining = total_steps - (i + 1)
        time_remaining = steps_remaining * time_per_step

        print(f'\rEpoch {epoch+1}/{num_epochs}, Step {i+1}/{total_steps}, Progress: {progress*100:.2f}%, Loss: {loss.item():.4f}, Time Remaining: {time_remaining:.2f} seconds', end='')

    epoch_accuracy = correct / total
    epoch_loss = running_loss / len(train_loader)

    epoch_time = time.time() - start_time  # Calculate epoch time
    # Update average epoch time
    avg_epoch_time = (epoch_time + avg_epoch_time * epoch) / (epoch + 1)
    # Print epoch-wise metrics including epoch time and estimated time for next epoch
    print(f'\rEpoch {epoch+1}/{num_epochs}, Loss: {epoch_loss:.4f}, Accuracy: {epoch_accuracy:.4f}, Time: {epoch_time:.2f} seconds, Expected Time for Next Epoch: {avg_epoch_time:.2f} seconds')

    # Update learning rate
    scheduler.step(epoch_loss)

    # Validate the model
    resnet.eval()
    val_loss = 0.0
    val_correct = 0
    val_total = 0
    with torch.no_grad():
        for data in test_loader:
            inputs, labels = data
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = resnet(inputs)
            val_loss += criterion(outputs, labels).item()
            _, predicted = torch.max(outputs.data, 1)
            val_total += labels.size(0)
            val_correct += (predicted == labels).sum().item()

    val_accuracy = val_correct / val_total
    val_loss /= len(test_loader)

    # Print validation metrics
    print(f'Validation Loss: {val_loss:.4f}, Validation Accuracy: {val_accuracy:.4f}')

    # Save the model if it has the best validation loss so far
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        torch.save(resnet.state_dict(), 'resnet_model.pth')
        print('Model saved.')

print('Finished Training')
```
